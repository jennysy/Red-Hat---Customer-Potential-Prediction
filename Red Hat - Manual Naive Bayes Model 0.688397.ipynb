```python
import pandas as pd
import numpy as np
import os
```

```python
file_path = 'C:\\hudsondata\\Machine Learning\\Redhat'
acttest = pd.read_csv(os.path.join(file_path,'act_test.csv'))
acttrain = pd.read_csv(os.path.join(file_path,'act_train.csv'))
people = pd.read_csv(os.path.join(file_path,'people.csv'))
```

```python
people = people.replace({'type':'','group':'','ppl_':''},regex=True)
acttrain = acttrain.replace({'type':'','group':'','ppl_':''},regex=True)
acttrain = acttrain.fillna('-1')
acttest = acttest.replace({'type':'','group':'','ppl_':''},regex=True)
acttest = acttest.fillna('-1')
```

```python
joined_train=pd.merge(people, acttrain, on='people_id')
joined_test=pd.merge(people, acttest, on='people_id')
```

```python
a=dict()
b=dict()
```

```python
#did not include char_38 because it is a continuous variable and will most likely decrease the predictive capability of naive bayes given its continuous form
#first - extract all data from 0 outcome
#second - get all the unique values from that column (regardless of outcome) and append to first extraction to make sure all values are represented
#third - given the outcome, get the probabilities of each value for each column
#fourth - get the negative log probability of each and store all in a dictionary, this way we allow to computer to work with manageable values (not super small)
# fifth - do the same for outcome 1
```

```python
for name in joined_train.columns.values:
    if name!= 'people_id' and name!= 'activity_id' and name!='date_x' and name!='date_y'and name!='char_38':
        print name
        a[name] = joined_train[joined_train['outcome']==0][name]
        a[name] = a[name].append(pd.DataFrame({0:joined_train[name].unique()}))[0]
        a[name] = -np.log(a[name].value_counts(dropna=False)/len(a[name]))
        a[name] = dict(a[name])
        b[name] = joined_train[joined_train['outcome']==1][name]
        b[name] = b[name].append(pd.DataFrame({0:joined_train[name].unique()}))[0]
        b[name] = -np.log(b[name].value_counts(dropna=False)/len(b[name]))
        b[name] = dict(b[name])
```

```python
aa = np.zeros(len(joined_test))
bb = np.zeros(len(joined_test))
```

```python
b
```

```python
for name in joined_test.columns.values:
    if name != 'people_id' and name != 'activity_id' and name != 'date_x' and name != 'date_y' and \
       name != 'char_38' and name != 'group_1' and name != 'char_10_y':
            print name
            aa = aa + np.array([a[name][x] for x in joined_test[name]])
            bb = bb + np.array([b[name][x] for x in joined_test[name]])
```

```python
#from all the inverse logarithm probability of every char (feature),we added everything up and we then use np.argmin to give us the smaller number between
# the two and whichever that is - that is our prediction - which is either 0 or 1
```

```python
submission=pd.DataFrame({'activity_id':acttest['activity_id'], 'outcome':np.argmin(np.concatenate((aa, bb)).reshape((2, len(aa))).T, 1)})
```
